# Plot Function

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = x**2
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, '+')
```

```python
plt.plot(x, y, 'v-')
```

```python
plt.plot(x, y, 'vr-')
```

```python
plt.plot(x, y, 'hg:')
```

```python
y2 = x+x**2
plt.plot(x, y, 'b-o', x, y2, 'r-o')
```

```python
plt.plot(x, y, "b-o")
plt.plot(x, y2, "r-o")
```

```python
plt.plot(x, y, "b-o")
```

```python
plt.plot(x, y2, "r-o")
```

```python
plt.plot(x, y, "green")
```

```python
plt.plot(x, y, color = 'blue', linestyle = '--', marker = 'o', fillstyle = 'none')
```
