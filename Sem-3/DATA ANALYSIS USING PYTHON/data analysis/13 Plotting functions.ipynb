# Plotting Functions

The `pyplot` sub-module provides several plotting functions, designed to create various types of standard plots.

## `plot` function

```python
from matplotlib import pyplot as plt
import numpy as np
```

### State-based approach

```python
x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.sin(x), label="Sine")
plt.plot(x, np.cos(x), label="Cosine")
plt.legend()
plt.title("Sine and Cosine")
plt.show()
```

### Object Oriented Approach

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label="Sine")
ax.plot(x, np.cos(x), label="Cosine")
ax.legend()
ax.set_title("Sine & Cosine")
plt.show()
```

In object oriented approach, since the objects are retained in memory, we can modify the objects and redraw.

```python
ax.set_title('Sine and Cosine')
fig
```

The `plot` function also accepts `data` argument to specify the dataset to be ploted. When `data` argument is used, `x` argument can be the string specifyng label of the data to be plotted on x-axis. Similarly for `y` argument.

```python
from pydataset import data
iris = data('iris')
plt.plot('Petal.Width', 'Petal.Length', ".", data = iris)
plt.grid()
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
plt.title("Petal Length against Petal Width")
plt.show()
```

We can distinguish between different specied as shown in the following code

```python
for sp in iris.Species.unique():
    plt.plot(iris['Petal.Width'][iris.Species == sp], 
             iris['Petal.Length'][iris.Species == sp], 
             ".", 
             label = sp)
plt.grid()
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
plt.title("Petal Length against Petal Width")
plt.legend()
plt.show()
```

Note, however, that the `data` argument is not used in above code. This is because we need to use expressions for `x` and `y` arguments.

### Using format strings

```python
x = np.linspace(-3, 3, 30)
y = x**2
plt.plot(x, y)  # Without format string
```

```python
plt.plot(x, y, '+')
```

```python
plt.plot(x, y, 'v-')
```

```python
plt.plot(x, y, 'vr-')
```

Any permutation of characters in format string will produce the same result.

```python
plt.plot(x, y, '-rv')
```

```python
plt.plot(x, y, 'hg:')
```

### Providing multiple triplets

The `plot` function also takes multiple triplets as arguments to plot multiple data sets.

```python
y2 = x+x**2
plt.plot(x, y, 'b-o', x, y2, 'r-o')
```

Alternatively, one can use the `plot` function call multiple times. 

```python
plt.plot(x, y, "b-o")
plt.plot(x, y2, "r-o")
```

One can explicitely specify the color, line style, marker etc by providing corresponding arguments.

```python
plt.plot(x, y, color = 'blue', linestyle = '--', marker = 'o', fillstyle = 'none')
```

## `boxplot` function

```python
z=np.random.randn(100)
plt.boxplot(z)
plt.show()
```

Horizontal orientation for a boxplot can be obtained as shown below.

```python
plt.boxplot(z, vert = False)
plt.show()
```

Multiple datasets can be plotted as shown below.

```python
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(5, 1.3, 100)
x3 = np.random.normal(3, 0.5, 100)
plt.boxplot([x1, x2, x3], labels = ['$X_1$', '$X_2$', '$X_3$'])
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.boxplot([x1, x2, x3], labels = ['$X_1$', '$X_2$', '$X_3$'])
plt.title("Comparing three datasets")
plt.show()
```

```python
style.use('dark_background')
plt.boxplot([x1, x2, x3], labels = ['$X_1$', '$X_2$', '$X_3$'])
plt.grid(linestyle = '--')
plt.title("Comparing three datasets", {'fontsize': 16, 'color': "yellow"})
plt.show()
```
