### 1. Generate a random sample size 100 from normal distribution with mean 12 and standard deviation 1.5.

```python
import numpy as np
random = np.random.normal(12,1.5,100)
print("100 random samples are:",random)
```

### 2. Create a series named 'X' containing this data with indices 'u1','u2',....'u100'.

```python
import pandas as pd
import numpy as np
k=pd.Series(random,index=['u'+str(i) for i in range(1,101)])
k.name='x'
print(k)
```

### 3.Consiering the contents of above series as a population , computing population mean and population standard deviation.

```python
print("mean of series:",k.mean())
```

```python
print("standard deviation of series:",k.std())
```

### 6. Select 50 samples , each of size 20 , without replacement from the population. Obtain sample means and sampl standard deviations of these 50 samples . Store these means and standard deviations in data frame with suitable indices.

```python
M=[]
S=[]
for i in range(50):
    i=np.random.choice(k,size=20,replace=False,p=None)
    mean=np.mean(i)
    M.append(mean)
    sd=np.std(i)
    S.append(sd)
d=[M,S]
d1=pd.DataFrame (d). T
d1.index=('s'+str(i) for i in range(1,51))
d1.columns=("mean","std")
d1
```
