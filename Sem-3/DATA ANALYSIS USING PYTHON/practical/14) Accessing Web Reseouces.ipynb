### Using data.gov.in API

To access data from `data.gov.in` platform using API, you need to create an account on the postal. Once an account is created, the API Key will be assigned to you. Use your API Key in the code given below.

```python
Key = '' # USe your own key here
url = 'https://api.data.gov.in/resource/daee9216-a2aa-4dc0-a2a4-81249823dc58?api-key='+Key+'&format=json&offset=0&limit=100'
resp = requests.get(url)
```

`resp` is the `Response` object containing the response received from the server.

since we have requested data in json format, we can use the response objectâ€™s json method to return a list containing JSON parsed into dict objects.

```python
data = resp.json()
data
```

```python
InstData = pd.DataFrame(data['records'])
InstData.columns
```

## Using Twitter API

```python
import tweepy
```

```python
# Use your own Keys and Tockens here
APIKey= '' 
APISecretKey = ''
AccessToken = ''
AccessTockenSecret = ''
```

```python
auth = tweepy.OAuthHandler(APIKey, APISecretKey)
auth.set_access_token(AccessToken, AccessTockenSecret)
```

```python
api = tweepy.API(auth)
query = '#Vadodara'
cursor = tweepy.Cursor(api.search_tweets, q = query, lang = "en")
```

```python
import json
tweets = []
for page in cursor.pages(): 
   for item in page:
      tweets.append(item._json)
outfile = open(r'd:\tweets.json', 'w')
json.dump(tweets,outfile)
outfile.close()
```
