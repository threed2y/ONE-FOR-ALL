## 1) Generate a population of real values containing 1000 different values.

```python
import random
import numpy as np
from numpy.random import default_rng
rng=default_rng()
```

```python
population=rng.integers(1,50,1000,endpoint=True)
population
```

## 2) Select 1000 samples each of size 50 with replacement and compute mean.

```python
import scipy as sp
from scipy import stats
```

```python
mean_2=[]
for i in range(1000):
    sample=rng.choice(population,50,replace=True,p=None)
    mean=sample.mean()
    mean_2.append(mean)
    print("mean of ",i+1,"sample with replacement is:",mean)
```

## 3) Select 1000 samples each of size 50 without replacement and compute mean.

```python
mean_1=[]
for i in range(1000):
    sample=rng.choice(population,50,replace=False,p=None)
    mean1=sample.mean()
    mean_1.append(mean1)
    
    print("mean of ",i+1,"sample without replacement is:",mean1)
```

## 4) For that 1000 mean compute mean and variance for with and without replacement and compare mean and variance of both and interprete it.

```python
import pandas as pd
d=[mean_2,mean_1]
df=pd.DataFrame(d,index=['Mean WR','Mean WOR']).T
```

```python
df.head()
```

```python
meanvar=df.agg(['mean','var'])
meanvar
```

#### Here mean of without replacement is slightly more than with replacement.

#### And variance of without replacement is less than with replacement.

## 5) For those 1000 mean check normality for with and without replacement.

```python
normwr=stats.normaltest(mean_2)
normwr
```

#### Conclusion: Here , pvalue is less than 0.05.Hence we reject null hypothesis and say that the mean of with replacement  is normally distributed.

```python
normwor=stats.normaltest(mean_1)
normwor
```

#### Conclusion: Here , pvalue is greater  than 0.05.Hence we donot reject null hypothesis and say that the mean of without replacement  is not  normally distributed.
