### 1. Load the Iris data in a DataFrame.

```python
from pydataset import data
iris=data('iris')
iris.head()
```

### 2.	Create GroupBy object using Species as the key. 

```python
byspecies=iris.groupby('Species')
```

### 3.	For each group, obtain following summary statistics:

##### (i) No of observation (ii) Mean (iii) Variance (iv) Skewness (iv) Kurtosis

```python
import pandas as pd
from pandas import DataFrame
byspecies.agg(['count','mean','var','skew',DataFrame.kurt])
```

## 4.	Iterate over the GroupBy object, obtaining histogram of Petal length for each group

```python
from matplotlib import pyplot as plt
for Species, data in byspecies:
    print('Histogram for petal length',Species)
    data['Petal.Length'].hist()
    plt.show()           ## plotted graph below is the histogram of species and petal length.
```

## 5.	Prepare histogram for all four variables by iterating over appropriately defined GroupBy object.

```python
for Species, data in byspecies:
    print('Histogram for ',Species,'is:')
    data.hist()
    plt.tight_layout()
    plt.show()  ## plotted below graphs are the hISTOGRAMS of species with all four variable.
```

## 6.	Use pivot table functionality to prepare pivot table with species defining the rows, computing mean and standard deviation for each cell of the table.

```python
iris.pivot_table(index='Species',aggfunc=['mean','std'])
```
