* For the Iris dataset, Consider random vector X = (Sepal Length, Sepal Width, Petal Length, Petal Width). Write Python code to perform the following tasks without using any readymade functions for performing tests of hypothesis.

1. Obtain unbiased estimators of mean vector and variance-covariance matrix for each of the three species.
2. Test whether the mean vector for Setosa is equal to (5, 3.5, 1.5, 0.25).
3. Test whether mean vector for versicolor is same as that of virginica.

* Now Perform tasks 2 and 3 using the library functions, and compare the results with the above results obtained by you.

```python
import pandas as pd
import numpy as np
from pydataset import data
iris = data("iris")
iris.head()
```

```python
setosa = iris[iris['Species'] =='setosa'].iloc[:,0:4]
versicolor = iris[iris['Species'] =='versicolor'].iloc[:,0:4]
virginica = iris[iris['Species'] =='virginica'].iloc[:,0:4]
```

## 1. Obtain unbiased estimators of mean vector and variance-covariance matrix for each of the three species.

## Unbiased estimators of mean vector for each Species

```python
setosa_mean = setosa.mean()
versicolor_mean = versicolor.mean()
virginica_mean = virginica.mean()
print("Unbiased estimator of Mean vector for setosa Species:", tuple(round(setosa_mean,5)))
print("Unbiased estimator of Mean vector for versicolor Species:", tuple(round(versicolor_mean,5)))
print("Unbiased estimator of Mean vector for virginica Species:", tuple(round(virginica_mean,5)))
```

## Unbiased estimators of var-cov vector for each Species : 

```python
n1 = setosa.count()
n2 = versicolor.count()
n3 = virginica.count()

var_setosa = n1/(n1-1) * setosa.cov()
var_versicolor = n2/(n2-1) * versicolor.cov()
var_virginica = n3/(n3-1) * virginica.cov()
print("Unbiased estimator of variance-covariance for setosa species: \n \n ",var_setosa)
print("Unbiased estimator of variance-covariance for versicolor species: \n \n ",var_versicolor)
print("Unbiased estimator of variance-covariance for virginica species: \n \n ",var_virginica)
```

## 2. Test whether the mean vector for Setosa is equal to (5, 3.5, 1.5, 0.25).
### ***One sample Hotelling T sqaure test***

## Hypothesis
#### H0 : µ1 = µ0 (Mean vector for setosa equals to (5,3.5,1.5,0.25))
#### H1 : µ1 ≠ µ0 (Mean vetor for setosa does not equals to (5,3.5,1.5,0.25))

```python
mu0 = np.array([5,3.5,1.5,0.25])
mu0 = mu0[:,np.newaxis]
Xbar = np.array(setosa_mean)
xbar = Xbar[:,np.newaxis]
S = (len(setosa)-1)* var_setosa
S
```

```python
Hott_Tsquare = len(setosa)*(len(setosa)-1)*((xbar-mu0).T @ (np.linalg.inv(S)) @ (xbar-mu0))
print(Hott_Tsquare)
Hott_tsquare = float(Hott_Tsquare)
print("Test statistics is:",Hott_tsquare)
```

```python
Fcal = (((len(setosa)-4)/4) * Hott_tsquare)/(len(setosa)-1)
Fcal
```

```python
from scipy import stats
## using table value at alpha = 0.05
Ftab = stats.f.ppf(q = 1 - 0.05, dfn=4, dfd=len(setosa)-4)
print("F-table value:",Ftab)
if Fcal < Ftab:
    print("We do not reject the H0.")
else:
    print("we reject the H0.")
```

```python
## using P -value
from scipy import stats
pvalue = stats.f.sf(Fcal, 4, 50-4)
print("p-value value:",pvalue)
if pvalue < 0.05:
    print("We reject the H0.")
else:
    print("we do not reject the H0.")
```

**Conclusion : Here, Fcal < Ftab  <u>or</u>  P-value > 0.05. So we do not reject null hypothesis and we can say that Mean vetor for setosa equals to (5,3.5,1.5,0.25).**

## 3. Test whether mean vector for versicolor is same as that of virginica.
### ***Two sample Hotelling T sqaure test***

## Hypothesis:
#### H0 : µ1 = µ2 (Mean vector for versicolor is same as that of virginica.)
#### H1 : µ1 ≠ µ2 (Mean vector for versicolor is not same as that of virginica.)


```python
S_inv = np.linalg.inv((var_versicolor*49+49*var_virginica)/98)
print(S_inv)
X1 = np.array(versicolor_mean)
print(X1)
X2 = np.array(virginica_mean)
print(X2)
```

```python
Tsq = ((50*50)/100)*((X1-X2).T) @ S_inv @ (X1-X2)
print("Test Statistics",Tsq)
```

```python
Fcal = ((50+50-4-1)/((50+50-2)*4))*Tsq
Fcal
```

```python
## using table value at alpha = 0.05
Ftab = stats.f.ppf(q = 1 - 0.05, dfn=4, dfd=50+50-4-1)
print("F-table value:",Ftab)
if Fcal < Ftab:
    print("We do not reject the H0.")
else:
    print("we reject the H0.")
```

```python
## using P -value
from scipy import stats
pvalue = stats.f.sf(Fcal, 4, 50+50-4-1)
print("p-value value:",pvalue)
if pvalue < 0.05:
    print("We reject the H0.")
else:
    print("we do not reject the H0.")
```

**Conclusion : Here, Fcal > Ftab  <u>or</u>  P-value < 0.05. So we reject null hypothesis and we can say that Mean vector for versicolor is not same as that of virginica.**
