## 6. Prepare boxplot of all four measurements, each plot showing three boxes corresponding to the three species.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(2,2))
ig, ax = plt.subplots(2, 2, 
                       constrained_layout = True)
plt.subplot(2,2,1)
df1.boxplot()
plt.title('Setosa')
plt.subplot(2,2,2)
df2.boxplot()
plt.title('Versicolor')
plt.subplot(2,2,3)
df3.boxplot()
plt.title('Virginica')

```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
df1.boxplot()
ax1.set_title('Setosa')
ax2=fig.add_subplot(2,2,2)
df2.boxplot()
ax2.set_title('Versicolor')
ax3=fig.add_subplot(2,2,3)
df3.boxplot()
ax3.set_title('Virginica')
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```
