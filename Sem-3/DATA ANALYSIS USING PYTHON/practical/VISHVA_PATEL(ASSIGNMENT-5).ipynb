### 1)Generate 1000 random samples from Normal distribution with mean 15 and variance 9. Test each of these samples for normality, using a suitable test of normality from the scipy library, and output percentage of the times the hypothesis of normality is rejected. Comment on your finding.


```python
import numpy as np
import scipy as sp
from scipy import stats
from numpy.random import default_rng
rng=default_rng()
norm15_3=stats.norm(loc=15,scale=3)
norm15_3

count_rej=0
count_accept=0
for i in range (1000):
    sample=stats.norm.rvs(loc=15,scale=3,size=50)
    test=stats.kstest(sample,norm15_3.cdf)
    if test.pvalue<0.05:
        count_rej +=1
    else:
        count_accept +=1
    
        
        
print("count_rej",count_rej)
print("count_accept",count_accept)

print("percentage of rejecting",round((count_rej/1000*100),4),"%")
print("percentage of acceptance",round((count_accept/1000*100),4),"%")
```

##### Hence we can conclude that 5.8% out of 1000 samples of normal distribution with mean 15 and variance 9 are being rejected

### 2) For the Iris dataset, test whether Petal length follows normal distribution for each of the species.

```python
from pydataset import data
iris = data("iris")

import pandas as pd
gbspecies = iris.groupby("Species")

for i,j in gbspecies:
    pvalue = stats.normaltest(j["Petal.Length"])[1]
    if pvalue < 0.05:
        print("we will reject null hypothesis for ",i,"that is ",i," data does not follow normal distribution")
    else:
        print("we will do not reject null hypothesis for",i,"that is",i,"data follows normal distribution")

```

##### 3)For the Iris dataset, test whether mean Petal length of 'Versicolour' species significantly differ from that of 'Virginica' species.


```python
VC = iris[iris.Species == "versicolor"]["Petal.Length"]
VN = iris[iris.Species == "virginica"]["Petal.Length"]
```

#### first we will test for equality of variance

```python
pvalue = stats.bartlett(VC,VN)[1]
if pvalue < 0.05:
    print("variance is unequal")
else:
    print("variance is equal")
pvalue
```

##### Because variance is equal we will test mean with t-test

```python
pvalue = stats.ttest_ind(VC,VN,equal_var = True)[1]
if pvalue < 0.05:
    print("mean Petal length of 'Versicolour' species significantly differ from that of 'Virginica' species")
else :
    print("mean Petal length of 'Versicolour' species significantly do not differ from that of 'Virginica' species")
```
