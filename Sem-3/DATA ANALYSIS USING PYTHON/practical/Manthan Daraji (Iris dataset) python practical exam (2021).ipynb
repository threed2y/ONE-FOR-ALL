```python
# (c) Dataframe containing (i) Mean (ii) Standard Deviation of Sepal length and Sepal Width 
```

```python
x = iris["Sepal.Length"]
```

```python
y = iris["Sepal.Width"]
```

```python
PL = iris["Petal.Length"]
```

```python
a = x.mean()
b = y.mean()
print("mean_sepal_length = " , a , 
      "mean_sepal_width = ", b)
```

```python
# Petal Length on Sepal Length
import statsmodels.formula.api as smf
no_intercept_model1=smf.ols("PL~x + y -1", iris).fit()
no_intercept_model1.summary()
```
