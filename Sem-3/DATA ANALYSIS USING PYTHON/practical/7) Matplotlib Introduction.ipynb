# Data Visualization using Matplotlib

Matplotlib provides a high-level interface for creating plots through the `pyplot` sub-module. The `pyplot` sub-module is designed to provide a MATLAB like user interface.

Apart from `pyplot` Matplotlib also provides another interface through its object oriented API, which provides more controls for creating customized plots.

In the present course We will focus only on the `pyplot` sub-module of Matplotlib package. 
This submodule is typically imported in python code as shown below.

```python
from matplotlib import pyplot as plt
```

Every plot created by Matplotlib resides in a `Figure` object. A `Figure` object, the top level object that contains all the plot elements, can be created as shown below.

```python
figA = plt.Figure()
```

```python
type(figA)
```

At this stage the figure object `figA` doesn't contain anything. 

```python
figA
```

To create a plot, we need to add it to the Figure object. In Matplotlib, a `subplot` object represents a plot. Thus, the next logical step in creating a plot is to add a `subplot` to the `Figure` object.

The `add_subplot` method is designed to add a `subplot` to the `Figure` object. It can be used as shown below.

```python
plot1 = figA.add_subplot()
```

```python
type(plot1)
```

At this stage, the Figure object `figA` contains a subplot object `plot1` with default axes.

```python
figA
```

Technically, a `subplot` class is derived from the class `Axes` as its base, proving some additional methods. In practice, we always use `subplot` objects instead of directly using the `Axes` objects.
