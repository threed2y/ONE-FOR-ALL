### 1. Generate a random sample of size 100 from normal distribution with mean 12 and standard deviation 1.5.

```python
import numpy as np
sample=np.random.normal(loc=12,scale=1.5,size=100)
sample
```

### 2. Crate a Series named 'X' containing this data with indices 'u1', 'u2', ..., 'u100'

```python
import pandas as pd
import numpy as np
k = pd.Series(sample, index = ['u' + str(i) for i in range(1,101)])
k.name='X'
k
```

### 3.Considering the contents of above series as a population, compute population mean  ùúá  and population standard deviation  ùúé .

```python
k.mean()            # population mean of above series
```

```python
k.std()             # population standard deviation of above series
```

### 6. Select 50 samples, each of size 20, without replacement from the population. Obtain sample means and sample standard deviations of these 50 samples. Store these means and standard deviations in a data frame with suitable indices.

```python
M=[]
S=[]
for i in range (50):
        i = np.random.choice(k, size=20, replace=False, p=None)
        mean = np.mean(i)
        M.append (mean)
        sd = np.std(i)
        S.append(sd)
d = [M,S]
d1 = pd.DataFrame (d). T
d1.index = ('s' + str(i) for i in range(1,51))
d1.columns= ('Mean', 'SD')
d1
```
