# Markdown Language

Given below are some examples of the use of Markdown language.

## <u> Headings </u>

# Heading 1
## Heading 2
### Heading 3
#### Heading 4

## <u> Bold, italics, ... </u>


This is an example of **bold** text  

This is a line containing *italics* text.

This is a line containing ***bold italics*** text.

This is a line containing `quoted` text.

This is a line containing ~~striked~~ out text.

## <u> Bulleted List </u>

* This is bulleted list item 1
* This is bulleted list item 2
* This is bulleted list item 3

Instead of using `*` characer in the Markdown code, , we can also use `+` OR `-` giving the same result.

## <u> Indented bulleted List </u>

* This is bulleted list item 1
  * This is indented bulleted list item 1
  * This is indented bulleted list item 2
    * Still inner level of bulleted list item
* This is bulleted list item 2

## <u> Ordered  List </u>

1. This is fist item of ordered list
1. This is second item of ordered list

This is a normal text

1. This is third item of ordered list
1. This is forth item of ordered list

*Remark: The actual number written in the code is not important. This insread of 1, 2, 3, the numbers could have been 1, 1, 1 OR 1, 6, 2 resulting in the same output.*

## <u> Indented Ordered  List </u> 

1. This is fist item of ordered list
  1. This is first line of indented ordered list

    2. This is second line of further indented ordered list 
  2. This is second line of indented ordered list

  * This is line 1 of bulleted list within ordered list

## <u> Writing Expressions </u>

Probability density function of normal distribution is 

$ f(x) = \frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{1}{2}\left (\frac{x-\mu}{\sigma}\right )^2} $

Here $\mu$ and $\sigma^2$ are the parameters of the distributions.
