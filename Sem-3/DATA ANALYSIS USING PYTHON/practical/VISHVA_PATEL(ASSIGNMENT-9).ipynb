```python
import pandas as pd
```

```python
opsd=opsd.asfreq('D')
opsd.Wind.isnull().any()
```

```python
opsd_Yearly_removed = opsd.rolling(365, center = True).mean()
fig, ax = plt.subplots()
ax.plot(opsd.Wind['Jan 2010':'Dec 2011'],
        marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(opsd_Yearly_removed.Wind['Jan 2010':'Dec 2011'],
        marker='.', linestyle='-', label='365-point Moving Average')
ax.set_ylabel('Wind Production (GWh)')
ax.legend();
```
