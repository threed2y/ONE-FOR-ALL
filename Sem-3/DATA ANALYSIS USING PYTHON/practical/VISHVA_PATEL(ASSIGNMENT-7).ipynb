## 1) Consider a quadratic model whose 50 values of x are to be generated from unifrom dist with 5 and 10 .

```python
import scipy as sp
from scipy import stats
```

```python
x=stats.uniform.rvs(5,10,50)
x
```

## 2) For Epsilon compute 50 values assuming it normally distributed with mean 0 and variance 16.

```python
Epsilon=stats.norm.rvs(0,4,50)
Epsilon
```

### Assuming beta(0)=10, beta(1)=2.5 and beta(2)= -0.5.
## 3)Now Compute Yvalues for that x values.

```python
y=10+2.5*x-0.5*x*x
y
```

## 4) Compute dataframe of x and y.

```python
import pandas as pd
d=[x,y]
df=pd.DataFrame(d,index=['x','y']).T
df
```

## 5) Fitting a linear model for the above dataframe.

```python
import statsmodels.formula.api as smf
model1=smf.ols("y~x+Epsilon",df)
result1=model1.fit()
result1.summary()
```

## 6) Plot residuals against x and observe the pattern and interprete it accordingly.

```python
from matplotlib import pyplot as plt
plt.plot(result1.resid,'x',linestyle='-',marker='')
```

## 7) Now add the quadratic term into the model and plot the residual against x value.Now observe the pattern and interprete accordingly.

```python
import statsmodels.formula.api as smf
model2=smf.ols("y~x+I(x*x)+Epsilon",df)
result2=model2.fit()
result2.summary()
```

```python
plt.plot(result2.resid,'x',linestyle='-',marker='')
```
