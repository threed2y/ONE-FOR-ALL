```python
print('keep on entering numeric values:')
total=0; count = 0
while True:
        x=input('enter a value:') 
        if x: # if the user supply any input, then if condition will be true
                try:
                        n = float(x)
                except:
                        print('Invalid Entry!')
                        continue
                total += n
                count += 1
        else: # if the user doesn't supply any input,then if condition will be false
                break
if count: # that is if count != 0
        print('count of numeric values = ',count)
        print('mean =',total/count)
else:
        print('no valid numeric value entered.')
```

```python
# Python’s built-in exec() function allows the user to execute arbitrary 
#Python code from a string or compiled code input.
# The exec() function can be handy when you need to run dynamically generated
#Python code, but it can be pretty dangerous if you use it carelessly. 
# Python’s built-in exec() function allows the user to execute any piece of
#Python code. With this function, you can execute dynamically generated code.
# That’s the code that you read, auto-generate, or obtain during your program’s
# execution. Normally, it’s a string.
# You can also use the user’s input or any other input source to construct 
#these strings.
# Once you’ve built the target code as strings, then you can use exec() to
#execute them as you would execute any Python code.
```

```python
# example:
# suppose we have written a certain code or a set of statements, 
# which we may wish to execute only under some situation
# so we define it as a string, and it will execute only when we will ask the 
# the interpreter to execute it using exec(name_of_string)
s="""
x=3
for i in range(4):
    print(i*x)"""
exec(s)
```

```python
"""
suppose that you have a sequence of numbers, and you want to build a new sequence 
containing the sum of squares of all the even numbers in an input sequence.

To solve this problem, you can use the following one-liner code:
"""
numbers = [2, 3, 7, 4, 8]
result=[number**2 for number in numbers if number % 2 == 0]
print("result =",result)
```

```python
"""
Enumerate object
The basic syntax is enumerate(sequence, start=0)

The output object includes a counter like so: 
(0, thing[0]), (1, thing[1]), (2, thing[2]),

As input it takes a sequence like a list, tuple or iterator. 
The start parameter is optional.
If the start parameter is set to one, counting will start from one
instead of zero

Create a sequence and feed it to the enumerate function. 
This can be any type of sequence, 
in this example we use a string/list. Then we output the object.
"""
```

```python
from datetime import datetime
d=datetime(1,1,1).now()
print(d)
print(d.strftime('%c'))
print(d.strftime('%x %X'))
print(d.strftime('It is %d day of the month'))
print(d.strftime('It is %j day of the year'))
print(d.strftime("Today's day is %a"))
print(d.strftime("Today's day is %A"))
print(d.strftime("This is %m month of the year")) 
print(d.strftime('This is %b month of the year')) # abbreviated name of month
print(d.strftime("This is %B month of the year")) # full name of month
print(d.strftime("This is %y year")) # year in 2 digits
print(d.strftime("This is %Y year")) # year in 4 digits
print(d.strftime("It's %H  hour")) # hour in 24 hrs format
print(d.strftime("It's %I%p hour")) # 12 hour format, am or pm
print(d.strftime("It's %M minutes")) 
print(d.strftime("It's %S seconds"))
```

```python
from datetime import datetime
from time import clock
d1=datetime(2011,2,1,7,33) # year,month,day,hour,min
d2=datetime(2017,10,19) 
print(d1.strftime('%x %X'))
print(d1.strftime('It is %d day of the month'))
print(d1.strftime("Today's day is %a"))
print(d1.strftime("This is %m month of the year")) 
print(d1.strftime('This is %b month of the year')) # abbreviated name of month
print("time elapsed since execution of the program so far=",clock())
print(d1.strftime("This is %Y year")) # year in 4 digits
print(d1.strftime("It's %H  hour")) # hour in 24 hrs format
print(d1.strftime("It's %M minutes")) 
if d1<d2:
    print(d2,"occurs after ",d1)
else:
    print(d1,"occurs after ",d2)
print("time elapsed since execution of the program so far=",clock())
```
