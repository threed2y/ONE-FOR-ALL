```python
import numpy as np
from scipy import stats
# Histogram
#syntax: random.normal(loc=0.0, scale=1.0, size=None)
sample=np.random.normal(size=1000)
classes=np.arange(-4,5)
print("bins array =",classes)
histogram=np.histogram(sample,bins=classes)
print("histogram:\n",histogram)

histogram_0=histogram[0]
freq_density=histogram_0/1000
print("freq density=",freq_density)
print("bins array=",histogram[1])
print(classes[1:])
print(classes[:-1])
bin_centers=0.5*(classes[1:]+classes[:-1]) # mid-points
print(classes[1:]+classes[:-1])
print(bin_centers)
pdf=stats.norm.pdf(bin_centers) # 'norm' is a distribution object
print("pdf=",pdf)
from matplotlib import pyplot as plt
plt.plot(bin_centers,freq_density,label="empirical distribution")
# this will plot the empirical (based on sample) distribution of data
plt.plot(bin_centers,pdf,label="theoretical distribution")
# this will plot the theoretical prob of std normal 
plt.legend()
plt.show()
```

```python
sample=np.random.normal(loc=5,scale=3,size=500) # mu=5,sigma=3
#print("sample=",sample)
#OR
# mu,sigma=5,3
# sample=np.random.normal(loc=mu,scale=sigma,size=50)
# or sample=np.random.normal(scale=sigma,size=50,loc=mu)
# or sample=np.random.normal(mu,sigma,50)
print("true mean=",5) 
print("true std dev=",3)
mean,std=stats.norm.fit(sample) # MLE of mu and sigma
print("MLE of mean=",mean)
print("MLE of std dev=",std)
```

```python
a = np.random.normal(20, 5, size=5000)
print("skewness of the data=",stats.skew(a))
print("kurtosis of the data, as per Fisher=",stats.kurtosis(a))
print("kurtosis of the data, as per Pearson=",(stats.kurtosis(a,fisher=False)))
# syntax: stats.kurtosis(a, fisher=True) #default value for parameter 'fisher' is True
# when fisher=True, 0 means mesokurtic curve
# when fisher=False, it gives Pearson's measure 
# of kurtosis, where the value will be 3 for 
# mesokurtic curve
```

```python
# Statistical Tests of Hypothesis
#-------------------------------
# One-sample t-test

a = np.random.normal(20, 5, size=100)
print(stats.ttest_1samp(a,18)) # test H0:mu=mu0=18
```

```python
## OR
t1=stats.ttest_1samp(a,18)
print("Output:\n",t1)
print("value of test statisitc=",t1[0]) # to extract value of test statisitc
print("p-value of the test=", t1[1]) # to extract p=value of the test
```

```python
# two-sample t-test for independent samples

a = np.random.normal(0, 1, size=100)
b = np.random.normal(1, 2, size=50)
t2=stats.ttest_ind(a, b)

#By default, equal_var = True
print("Output:\n",t2)
```

```python
t3=stats.ttest_ind(a, b, equal_var = False)# Welch-t-test
print("Output:\n",t3)
```

```python
# Paired t-test  

a = np.random.normal(0, 1, size=10)
b = np.random.normal(1, 1, size=10)
t4=stats.ttest_rel(a, b) # 'rel' means related samples
print("Output:\n",t4)
```

```python
# Non-parametric tests

x=np.array([6,8,9,2,3,4,6,7])
y=np.array([7,9,8,10,6,11,5,4])
z=np.array([12,15,18,14,19,10,11,9])
t5=stats.wilcoxon(x,y) # paired data
print("output:\n",t5)
t6=stats.mannwhitneyu(x,y) # independent samples
print("output:\n",t6)
t7=stats.kruskal(x,y,z) # more than two independent samples
print("output:\n",t7)
```
