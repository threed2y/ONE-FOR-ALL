```python
# use of if
n = int(input("Enter a number:"))
if n % 2 == 0:
    print(n," is an even number")
```

```python
# use of if-else
n = int(input("Enter a number:"))
if n % 2 == 0:
    print(n," is an even number")
else:
    print(n," is an odd number")
```

```python
from math import sqrt
n = int(input("Enter a number:"))
if n % sqrt(n) == 0:
    print(n," is a perfect square")
else:
    print(n," is not a perfect square")
```

```python
from random import randint
count = 0 # initialising 'count' with 0
for i in range(10):
    n=randint(1,15)
    print("the new integer selected randomly = ",n)
    if n % 3 == 0:
        count = count+1
print("No of integers which are multiple of 3 = ", count)
```

```python
# "elif" statement : multiple options / multiple courses of action

count1=0
count2=0
count3=0
count4=0

# OR count1 = count2 = count3 = count4 = 0  
# OR count1 = 0 ; count2 = 0 ; count3  =0 ; count4 = 0

for i in range(1,11):
    n = int(input("Enter an integer between 0 and 10 :"))
    if n == 0:
        count1 += 1  # is same as : count1 = count1 + 1
    elif n % 2 == 0:
        count2 += 1
    elif n % 3 == 0:
        count3 += 1
    else:
        count4 += 1
print("Number of zeroes inputted = ", count1)
print("Number of even values inputted = ", count2)
print("Number of values inputted which are multiple of 3, but not an even number = ", count3)
print("Number of nonzero values which are neither even nubmer nor a multiple of 3 = ", count4)
```

```python
count=0
for i in range(5):
    ch=input("enter an alphabet in smallcase:")
    if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u':
        print("inputted character is a vowel")
        count += 1
    else:
        print("inputted character is a consonant")
print("totalnumber of vowels inputted=",count)    
```

```python
from random import randint
count=0
for i in range(10):
    outcome=randint(0,1)
    if outcome == 0:
        print("It's a TAIL!")
    else:
        print("It's a HEAD !!")
        count += 1
print("number of heads in 10 tosses of a coin =",count)
    
```

```python
score=int(input("enter your score:"))
if score > 100 or score < 0:
    print('Invalid Score')
elif score >= 90 and score <= 100:
    print('Grade: A')
elif score >= 80:
    print('Grade: B')
elif score >= 70:
    print('Grade: C')
elif score >= 60:
    print('Grade: D')
elif score >= 40:
    print('Grade: E')
else:
    print('Grade: F')
```

```python
a1=float(input('enter a number:'))
a2=float(input('enter another number:'))
print("initially: a1= ",a1," , a2= ",a2)

temp=a1
a1=a2
a2=temp
print("After swapping: a1= ",a1," , a2= ",a2)
```

```python
# simple way to swap items in python
a1=float(input('enter a number:'))
a2=float(input('enter another number:'))
print("initially: a1= ",a1," , a2= ",a2)
a1,a2 = a2,a1
print("After swapping: a1= ",a1," , a2= ",a2)
```
