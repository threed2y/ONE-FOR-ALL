```python
animals=['cat','lion','giraffe','goat','tiger']
l1=[]
l2=[]
for a in animals:
    if len(a)== 4:
        l1.append(a)
    else:
        l2.append(a)
print("The animal names with number of characters equal to 4 are:")
print(l1)
print("The animal names with number of characters other than 4 are:")
print(l2)
```

```python
# ungrouped frequency distribution
from random import randint
l=[]
for i in range(1,51):
    r=randint(1,5)
    l.append(r)
    #l.append(randint(1,5))
print("list of 50 random numbers genereated between 1 to 5 is: ",l)
freq=[]
for i in range(1,6):
    f=l.count(i) 
    freq.append(f) # freq.append(l.count(i))
print("values : [1, 2, 3, 4, 5]")
print("freq= ",freq)
print("the frequency distribution is as follows:")
print("  x   freq ")
print("____________")
for i in range(1,6):
    print(i,"\t",freq[i-1])
      
```

```python
# frequency distribution for ungrouped data

#(when x takes all values within a range)
f=[]
L=[5,2,4,2,5,4,6,7,5,8,7,9,5,6,7,4,8,6] # assuming only integer values for data
for i in range(min(L),max(L)+1):
	freq=L.count(i)
	f.append(freq)
print('freq=',f)
print('The frequency distribution is as follows:')
print('x\tfreq')
print('----------------------')
for i in range(min(L),max(L)+1):
    print(i,"\t",f[i-min(L)])
```

```python
# frequency distribution for ungrouped data
#(when x does not take all values within a range)
f=[]
L=[1,4,3,5,4,6,7,5,7,9,5,6,7,4,3,6]
j=0
print('The frequency distribution is as follows:')
print('x\tfreq')
print('----------------------')
for i in range(min(L),max(L)+1):
        freq=L.count(i)
        if freq != 0:
                f.append(freq)
                j += 1
                print(i,"\t",f[j-1])
print('----------------------')
```
