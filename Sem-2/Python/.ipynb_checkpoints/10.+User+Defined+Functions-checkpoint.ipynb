```python
A function is a set of statements that take inputs, 
do some specific computation, and produce output. 
The idea is to put some commonly or repeatedly done tasks 
together and make a function so that instead of writing the 
same code again and again for different inputs, we can call the function.
Functions that readily come with Python are called built-in functions. 
Python provides built-in functions like print(), input() etc. 
But we can also create your own functions. These functions are known as user defined functions.
Syntax:
def function_name(comma separated list of parameters):
    statement block
```

```python
def greet():  # function header
    print("Hello Everyone!")

greet() # function call, that is, calling function 'greet'
```

```python
def greet(n): # function 'greet' has 1 parameter, n is dummy parameter/argument
    print("Hello Everyone!"*n)  
greet(2) # 2 is actual parameter
```

```python
m=int(input("enter an integer m: "))    
greet(m) # it will execute 'greet(n)' by replacing the value of n as m
```

```python
def f1(str,n=2): # here 2 is a default value of parameter 'n'
    print(str*n)
f1("hello",3) # call to function 'f1' with str='hello' and n=3
f1('good morning!') # call to function 'f1' with str='good morning!' and
# without specifying the value of n, but it is valid as the default 
# value of 'n' is 2
f1(n=4,str='hi ') # changing the parameter sequence is permissible 
# provided the parameter names are specified, 
# this is known as "calling function with keyword arguemnts" 
f1(4,'hi') # also permitted because both parameters types are different
```

```python
# function to convert temperature from degree celcius to degree fehrenheit
def temp(t):
    t1=t*9/5+32
    return t1 

f=temp(20)
print("f=",f)
print("the ",20," degree celcius =",f," degree fehrenheit")
t2=float(input("enter a temperature in degree celcius, t2:"))
f2=temp(t2)
print("the ",t2," degree celcius =",f2," degree fehrenheit")
t3=float(input("enter a temperature in degree celcius, t3:"))
print("the ",t3," degree celcius =",temp(t3)," degree fehrenheit")
```

```python
# when we expect some value from a function call, then such 
# functions should have a return statement, like function 'temp'
# but a function like 'greet' doesn't require any value in 
# return, because it is meant for only printing some message
# so, a return statement is not reuired for 'greet' function
```

```python
# function requiring multiple values to be returned
# multiple values will be returned by a function as a 'list'
from math import pi
def circle(r):
    area=pi*r*r
    circum=2*pi*r
    return[area,circum]  # if you write "return area,circum", then the output will be a tuple of 2 values
r1=float(input('enter radius:'))
l=circle(r1) # 'l' is a list object
print("list l =",l)
print('for a circle with radius',r1, 'the area=',l[0])
print('for a circle with radius',r1, 'the circumference= ',l[1])
```

```python
from math import pi
def circle(r):
    area=pi*r*r
    circum=2*pi*r
    return area,circum  # if you write "return area,circum", then the output will be a tuple of 2 values
r1=float(input('enter radius:'))
ans=circle(r1) # will be a tuple
print("area and circumference =",ans)
print('for a circle with radius',r1, 'the area=',ans[0])
print('for a circle with radius',r1, 'the circumference= ',ans[1])
```

```python
# function with default arguments
def f1(value=5 , repeat=2):
    v=str(value)
    print(v*repeat)
f1()
f1(4,3)
f1('@',4)
f1("hi!") # same as f1(value="hi!")
f1(repeat=5)
```

```python
def f2(a=2,b=5):
    print(a+b)
    print(a/b)
f2() # here a=2 and b=5 (both default)
f2(3) # here a=3 and b=5 (default)
f2(b=4) # here a=2 (default) and b=4
f2(10,15) #  if we want b=10 and a=15, this would be incorrect
f2(b=10,a=15) #  if we want b=10 and a=15, this would be correct
# Important: if we want to change the sequence of parameters while 
# calling a function, then it is must to speicfy the names of 
# the parameters, otherwise if we do not specify the parameter names
# then the parameters will be taken in sequence as specified in
# the function definition
```

```python
# function to compute simple interest
def Int(amount,rate,period):
    interest=amount*rate*period/100
    return interest
I=Int(10000,7.5,3)
print("I=",I)
#I1=Int(rate=7.5,amount=20000,years=5) # error because name of parameter is not as per the fun def
I1=Int(rate=7.5,amount=10000,period=3)
print("I1=",I1)
I2=Int(7.5,10000,3) # not appropriate way in general
print("I2=",I2)
```

```python
def factorial(n):
    prod=1
    if n == 0:
        return prod
    else:
        for i in range(1,n+1):
            prod = prod*i
        return prod
m=int(input('enter a positive integer:'))
fact=factorial(m)
print(m,"! =",fact)
```

```python
import sys
def factorial(n):
    if n<0:
        print("invalid n.")
        sys.exit()
    elif n == 0:
        return 1
    else:
        f = 1
        for i in range(1,n+1):
            f = i * f
        return f
m=int(input('enter a positive integer:'))
fact=factorial(m)
print(m,'! =',fact)
```
