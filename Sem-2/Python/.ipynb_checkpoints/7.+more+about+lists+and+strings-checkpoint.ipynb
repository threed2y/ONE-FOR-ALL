```python
f=[]
L=[1,4,3,5,4,6,7,5,7,9,5,6,7,4,3,6]

print('The frequency distribution is as follows:')
print('x\tfreq')
print('----------------------')
for i in range(min(L),max(L)+1):
        freq=L.count(i)
        f.append(freq)
        if freq != 0:
            print(i,"\t",f[i-min(L)])
print('----------------------')
```

```python
# frequency distribution for grouped data
from random import sample
n = int(input("Enter n :"))
L = sample(range(100),n) # a list of n observations between 0 to 99 will be generated 
"""
L=[]
for i in range(n):
    x=float(input("Enter an observation: "))
    L.append(x)
print("the data is:\n",L) 
"""
R = max(L) - min(L)
print("range = ",R)
k = int(input("number of classes:"))
h = float(input("class-width:"))
ll = [0 for i in range(k)]
ul = [0 for i in range(k)]
f=[0 for i in range(k)]
ll[0]=min(L)
for i in range(k):
        ul[i] = ll[i]+h
        if i < k-1:
                ll[i+1]=ul[i]
for i in L:
        for j in range(k):
                if i>=ll[j] and i<ul[j]:
                        f[j] += 1
                        break
                
print('The grouped frequency distribution is as follows:')
print('classes\t\t\tfreq')
print('--------------------------')
for i in range(k):
        print(ll[i]," - ",ul[i],"\t\t",f[i])

print('--------------------------')
print("\t\ttotal=",sum(f))
```

```python
date="21-04-2021" # date format is: day-month-year
d=date.split('-') # here '-' is a delimiter
# By default, the delimiter in 'split' will be a blank space
print(date)
print("d=",d)
d1=[int(i) for i in d]
print("d1=",d1)
```
