```python
n = int(input("enter an integer:"))
flag=0 # indicator for n is prime
for i in range(2,n):
    if n%i == 0:
        flag=1 # indicator for n is composite
        break
if flag == 0:
    print(n," is prime")
else:
    print(n," is composite")
```

```python
# to avoid printing the comma after the last character
st2='hello everyone!'
n=len(st2)
for ch in range(n):
    if ch<n-1:
        print(st2[ch],end=',')
    else:
        print(st2[ch])
```

```python
# changing the string characters
# Incorrect way
st3="hello evaryooe!"
st3[8]='e'
st3[12]='n'
print(st3)
```

```python
# Correct way
st3="hello evaryooe!"
print("original string:",st3)
st3=st3[0:8]+'e'+st3[9:12]+'n'+st3[13:]
print("modified string:", st3)
```
