```python
# Sending secret messsage - cipher text

original='abcdefghijklmnopqrstuvwxyz'
new='xznlwebgjhqdyvtkfuompciasr' # cipher code is fixed
message=input('enter the message:')
message=message.lower()
print('the cipher code of original message is:')
for ch in message:
    if ch.isalpha():
        print(new[original.index(ch)],end='')
    else:
        print(ch,end='')
```

```python
original='abcdefghijklmnopqrstuvwxyz'
new='xznlwebgjhqdyvtkfuompciasr' # cipher code is fixed
message=input('enter the message:')
message=message.lower()
print('the cipher code of original message is:')
s1=''
for ch in message:
    if ch.isalpha():
        s1=s1+new[original.index(ch)]
    else:
        s1=s1+ch
print("the encoded message is:",s1)
print("the original message, after decdoing, is:",end=' ')
s2=''
for ch in s1:
    if ch.isalpha():
        s2=s2+original[new.index(ch)]
    else:
        s2=s2+ch
print(s2)
```

```python
original='abcdefghijklmnopqrstuvwxyz'
import random
import string
# initializing an empty string and cipher the text using a single line code
# which is called 'string comprehension', using 'join'
new=''.join([random.choice(string.ascii_lowercase) for i in range(26)])
print('cipher text=',new)
```

```python
# another version to avoid duplication of alphabets while randomization
original='abcdefghijklmnopqrstuvwxyz'
import random
import string
new=''; s1=''
print("original sequence: ",original)
while True:
    s1= random.choice(string.ascii_lowercase)
    if s1 in new:
        continue
    new = new+s1  # use of '+' in place of 'join()'
    if len(new) == 26:
        break
print('cipher text=',new)

message=input('enter the message:')
message=message.lower()
print('the cipher code of original message is:')
for c in message:
    if c.isalpha():
        print(new[original.index(c)],end='')
    else:
        print(c,end='')
```
