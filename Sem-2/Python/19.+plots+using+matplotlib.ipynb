```python
# Bar Diagram
year=np.array([2010,2011,2012,2013,2014])
boys=np.array([50,60,65,80,75])
#plt.bar(year,boys)

plt.bar(year,boys,facecolor="yellow",width=0.5,
       edgecolor=np.array(['r','r','r','r','r']),linewidth=np.array([2,2,2,2,2]))

col=np.full(5,'b') # will generate an array of size 5, with all elements taking value 'r' 
lwd=np.full(5,3) # will generate an array of size 5, with all elements taking value '2'
plt.bar(year,boys,facecolor="yellow",width=0.5,
       edgecolor=col,linewidth=lwd)

plt.show()
```

```python
# Multiple Bar diagram
X = ['Group A','Group B','Group C','Group D']
Ygirls = [40,20,20,10]
Zboys = [30,30,25,20]
X_axis = np.arange(len(X)) # this is the array ([0,1,2,3])
plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')
  
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend() #OR plt.legend(loc='best') 
plt.show()
```

```python
# Horizontal bar diagram

dept=np.array(['D1','D2','D3','D4','D5'])
boys=np.array([50,60,65,80,75])
plt.barh(dept,boys,facecolor="yellow",edgecolor='b')
plt.xlabel("no of boys")
plt.ylabel("department")
plt.title("Number of Boys Students in each dept")
plt.show()
```

```python
# multiple line diagram, adding legends

fig=plt.figure()
ax=fig.add_subplot(1,1,1)
from numpy.random import randn
ax.plot(randn(1000).cumsum(),'k',label='first plot')
ax.plot(randn(1000).cumsum(),'g--',label='second plot')
ax.plot(randn(1000).cumsum(),'r.',label='third plot')
ax.legend(loc='best')
ax.set_xlabel('obs no')
ax.set_ylabel('cum sums')
ax.set_title('Multiple Line Diagram')
plt.show()
```

```python
# to add shape to a plot

fig=plt.figure()
ax=fig.add_subplot(1,1,1)
rect=plt.Rectangle((0.2,0.75),0.4,0.15,color='r',alpha=0.8)

# vertex=(0.2,0.75),length=0.4,height/width=0.15
circ=plt.Circle((0.7,0.2),0.15,color='b',alpha=0.4)
# centre=(0.7,0.2),radius=0.15
pgon=plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],
                color='g',alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6), dpi=80)
# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
# Set x limits
plt.xlim(-4.0, 4.0)
# Set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# Set y limits
plt.ylim(-1.5, 1.5)
# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
# Save figure using 72 dots per inch
# plt.savefig("exercise_2.png", dpi=72)
plt.show()
```

```python
# 3-d plots
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining all 3 axis
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```

```python
# importing mplot3d toolkits
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x + y
ax.scatter(x, y, z, c = c)
 
# syntax for plotting
ax.set_title('3d Scatter plot')
plt.show()
```
