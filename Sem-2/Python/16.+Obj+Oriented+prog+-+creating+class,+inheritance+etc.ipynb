```python
# object oriented programming
# object:- consists of 2 things- data and functions (methods)
# the functions work with the data
# any entity defined is an object, eg, string, list, variable
# eg, if we define a string: str1="hello"
# 'str1' is an object of 'string' class
# 'string' class has 2 things:- data and functions (or methods)
# eg., 'string' class now consists of a data with name 'str1'
# and the methods (functions) can now be defined for object 'str1'
# some other examples of class are: 'int','float','list',
# 'tuple','set','dictionary' etc.

# It is possible for the user to define his/her own class
# if the user wishes to have a data which doesn't fit
# into any of the standard exisitng data types (class) 

# 'class' is a template for objects. It contains code for all the methods
# pertaining to that 'object'

# Example: we will define a new class named 'point'
# we actually want to work with a point in 2-dimension
# with x-coordiante and y-coordinate, like (x,y)
# eg. a point (0,0) represents an origin

# eg. to define a string class:
# class string(object):
 #     def __init__():
 #         ----
 #         ----
```

```python
Class Variables
These are the variables that are shared across all instances of a class.
It is defined at the class level, outside any methods. 
All objects of the class share the same value for a class 
variable unless explicitly overridden in an object.

Instance Variables
Variables that are unique to each instance (object) of a class. 
These are defined within __init__ method or other instance methods.
Each object maintains its own copy of instance variables, 
independent of other objects.

For example, in the next program, 
```

```python
class Model:
    # Class variable
    Type = "Reg"

    def __init__(model, name, regressor):
        # Instance variables
        model.name = name
        model.regressor = regressor

# Create objects
model_1 = Model("Simple", 1)
model_2 = Model("Multiple", 3)

# Access class and instance variables
print(model_1.Type)  # (Class variable)
print(model_1.name)     # (Instance variable)
print(model_2.name)     # (Instance variable)

# Modify instance variables
model_1.name = "simple regression model"
print(model_1.name)     # (Updated instance variable)

print(model_2.name) 
# Modify class variable
Model.Type = "linear reg"
print(model_1.Type)  # (Updated class variable)
print(model_2.Type)  
```

```python
Explanation:

Class Variable (species): 
Shared by all instances of the class. Changing Model.Type affects all objects,
as itâ€™s a property of the class itself.

Instance Variables (name, regressor):
Defined in the __init__ method. Unique to each instance 
(e.g., model_1.name and model_2.name are different).
Accessing Variables: Class variables can be accessed via the 
class name (Model.Type) or an object (model_1.Type).

Instance variables are accessed via the object (model_1.name).
Updating Variables: Changing Model.Type affects all instances.
Changing model_1.name only affects model_1 and does not impact model_2.
```

```python
class Point:
    def __init__(pt,x=0,y=0):
        pt.x=x
        pt.y=y
    def __str__(p1):
        return"({0},{1})".format(p1.x,p1.y)
class Rectangle:
    def __init__(r,c,w,h):
        r.corner=c
        r.width=w
        r.height=h
    
    def __str__(r):
        return "({0},{1},{2})".format(r.corner,r.width,r.height)
    def grow(r,delta_w,delta_h):
        r.width +=delta_w
        r.height += delta_h
    def move(r,dx,dy):
        r.corner.x += dx
        r.corner.y += dy
R1=Rectangle(Point(10,5),100,50)
print(R1)
R1.grow(25,-10)
print(R1)
R1.move(-10,10)
print(R1)
```

```python
def r_center(r): # it is a user-defined function but not a method for 'ractangle' class, also 'r' is a 'rectangle' object
        p=Point()
        p.x = (r.corner.x + r.width)/2.0
        p.y = (r.corner.y + r.height)/2.0
        return(p)
    
box1=Rectangle(Point(0,0),100,200)
print("box1:",box1)

ct1=r_center(box1)
print("the center of rectangle is:")
print(ct1)

box2=Rectangle(Point(20,30),100,200)
print("box2:",box2)

ct2=r_center(box2)
print("the center of rectangle is:")
print(ct2)
```

```python
# Concept of "inheritance"
class Parent: # can be called a 'base class'
    def __init__(x,a): #'a' must be a string
        x.a=a
    def method1(x):
        print(x.a*2)
    def method2(x):
        print(x.a+'***')
class Child(Parent): # this means 'Child' class is inherited from 'Parent' class
    def __init__(x,a,b): # a and b must be strings
        x.a=a
        x.b=b
    def method1(x): # 'method1' of the 'child' class will override 'method1' of the 'parent' class
        print(x.a*7)
    def method3(x):
        print(x.a+x.b)
p=Parent('Hi')
c=Child('OK','Bye')
print('method1 of parent class:')
p.method1()
print('method2 of parent class:')
p.method2()
print('method1 of child class:')
c.method1()
print('method2 of child class:')
c.method2()
print('method3 of child class:')
c.method3()
```

```python
# concept of "Polymorphism"
# A function that can work with different types of objects
# may be called a "polymorphic" function
# The function "histogram" which we are defining below 
# can be called "polymorphic" because it can work for
# a 'string' object, a 'list' object or a 'tuple' object or a 'set' object

def histogram(s): 
    d=dict()
    for c in s:
        if c not in d:
            d[c]=1
        else:
            d[c]+=1
    return(d)
s1='have a nice day!!'
h1=histogram(s1)
print("Dictionary consisting of no of occurences of each character in given string s:")
print(h1)

t=['x1','x2','x1','x1','x3','x2']
h2=histogram(t)
print("Dictionary consisting of no of occurrences of each item of the list:")
print(h2)

t1=(1,2,3,4,3,2,4,1,2,2)
h3=histogram(t1)
print("Dictionary consisting of no of occurrences of each item of the tuple:")
print(h3)

t3=[4,6,5,8,7,6,5,4,5,6,7,7,6,8,5,6,5,6]
h4=histogram(t3)
print("Dictionary consisting of no of occurrences of each value of the list:")
print(h4)

s2={4,6,5,8,7,6,5,4,5,6,7,7,6,8,5,6,5,6}
h5=histogram(s2)
print("Dictionary consisting of no of occurrences of each value of the set:")
print(h5)       
```
