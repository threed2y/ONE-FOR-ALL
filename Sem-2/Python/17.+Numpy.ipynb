```python
b1=np.random.rand(4) 
# will generate an array of 4 uniform random numbers in range [0,1]
print("b1=",b1)
b2=np.random.randn(5)
# will genertate an array of 5 random numbers from N(0,1)
print("b2=",b2)
# we know, if X~N(mu,sigma), then z=(X-mu)/sigma ~ N(0,1)
# Then X = mu +sigma*z, where Z is a std normal
b3=2.5*np.random.randn(5)+3
# will generate an array of 5 random numbers from N(mean=3,sd=2.5)
print("b3=",b3)
b4=2.5*np.random.randn(2,4)+3
# will generate 2x4 array of random numbers from N(mean=3,sd=2.5)
print("b4=",b4)
b5=np.random.randint(0,21,10)
# will generate an array of 10 random numbers between 0 and 21
print("b5=",b5)
```

```python
# basic operations with array
# -------------------------------------
# Arithmetic and relational operations
print(b5%3==0)
# will create an array of logical values
print(b5[b5%3==0])
# will extract those elements of b5 which are
# evenly divisible by 3
b5[b5%3==0]=-1
# will replace those elements of b5 which are evenly
# divisible by 3, by -1
print("b5=",b5)
print(b5[[2,3,5,7]])
# will print elements with specified indices of b5 as an array
```

```python
a=array([1,2,3,4]) # same as a=np.array([1,2,3,4])
print(a+1)
print(2**a)
b=np.ones(4)+1
print("b=",b)
print(a-b)
a1=array([[1,2],[3,4]]) # same as a1=np.array()
print("a1=",a1)
b1=array([[2,3],[4,5]])
print("b1=",b1)
print("a1+b1=",a1+b1) # element-wise addition
add=np.add(a1,b1) # element-wise addition (same as a1+b1)
print("addition of a1 and b1=",add)
print("a1-b1=",a1-b1)  # element-wise subtraction
sub=np.subtract(a1,b1) # element-wise addition (same as (a1-b1))
print("element wise subtraction a1-b1=",sub)
print("element wise multiplication:",a1*b1) # OR USE - np.multiply(a1,b1)
# In element-wise multiplication, both arrays should be of same dimension
print("matrix multiplication:",a1.dot(b1)) # dot product, that is, usual matrix multiplication
print(a1/b1) # division , alternatively, np.divide(a1,b1)
```

```python
# use of relational operators
import numpy as np
c1=np.array([1,2,3])
c2=np.array([2,3,3])
print(c1==c2)
# The above code will do element-wise comparison and return T / F 
# corresponding to each element
#c3=array([2,3,3,1])
#print(c1==c3) # will be an error because of unequal length

print(c1>c2) # element-wise comparison
# print(c1<c3) will be an error because of unequal length

print(np.array_equal(c1,c2)) # will return True or False, only one value
# as a final answer, not element-wise comparison

c4=np.array([3,1,2])

print(np.array_equal(c1,c4))
c5=np.array([1,2,3])
print(np.array_equal(c1,c5))
```

```python
# logical operations
d1=np.array([True,False,True,False])
d2=np.array([False,False,True,True])
print(np.logical_or(d1,d2))
print(np.logical_and(d1,d2))
print(np.logical_not(d1))
d3=np.array([5>4,10<2])
d4=np.array([9%3,10%4])
print(d3)
print(d4)
print(np.logical_or(d3,d4))
print(np.logical_and(d3,d4))
print(np.logical_not(d3))
```
