```python
# Concept of Namespace: namespace is a collection of identifiers
# that belong to a  module or a function.
# Each module/function has its own namespace.
# Thus, we can use same identifier names in multiple modules
# There is a one-to-one correspondence / mapping between a
# module/function and a namespace
# A single module should be defined as a single file
# which gives rise to one namespace
# python takes the module-name as a file-name and this becomes 
# the name of the namespace
# Eg. "math" module is defined as filename "math.py", 
# the corresponding module name is "math", and its namespace
# is "math"

# The file named as "module2.py", which is defining module 
# named "module2", its namespace will be "module2", and it has
# two identifiers, 'question' and 'answer' 

# The file named as "module3.py", which is defining module 
# named "module3", its namespace will be "module3", and it has
# two identifiers, 'question' and 'answer' 
```

```python
import module2
import module3
print(module2.question)
print(module2.answer)
#print(module3.question)
#print(module3.answer)
f1=module3.f() # f1 will be a list
print(f1[0])
print(f1[1])
```

```python
# Example to demonstrate that functions have
# their own namespaces
def f():
    n=7 # this n is in namespace 'f'
    print("value of n in function f is:",n)
    
def g():
    n=17 # this n is in namespace 'g'
    print("value of n in function g is:",n)
    
n=5 # this n is in namespace "program_name"
print("value of n outside f and g is:",n)
f()
g()
```

```python
## File handling: 
Opening a file to read and / or write data:
# opening a file creates, what is called, 'a file handle'
# If the file to be created in different directory,
# then we need to specify the path

In the code given below, 
# 'myfile' is an identifier which refers to a 'new handle object'
# 'w' is a mode for writing
# 'write' is a method on the 'handle' (myfile)
# this statement will 'invoke' the 'write' method on the 
# handle 'myfile'
```

```python

myfile=open("D://test.txt","w") 
myfile.write("first file written from Python\n")
myfile.write("another way to write to the file is as below:\n")
line1="Good Morning!\n"
myfile.write(line1)
myfile.close() # this is a 'close' method for file handle
```
