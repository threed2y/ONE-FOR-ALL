```python
s1={1,2,3}
s2={3,4}
print("union of s1 and s2 is: ",s1|s2) # union of two sets; '|' operator
print("intersection of s1 and s2 is: ",s1&s2) # intersection of two sets; '&' operator
print("difference between sets s1 and s2 is: ",s1-s2) # operator '-'
print("symmetric difference between s1 and s2 is: ",s1^s2)
# symmetric difference between s1 and s2 = (s1 union s2)-(s1 intersection s2)
print("check whether 3 is in s1 ?:",3 in s1) # use of 'in' operator
print("check whether 2 is not in s1 ?: ", 2 not in s1)
```

```python
# 'set' methods
# suppose 's' is a set
# s.add(x) - will add element 'x' to set 's' 
s={1,3,4,6,8}
s1=s.copy()
print("set s= ",s)
print("set s1= ",s1)
s.add(2)
print("set s= ",s)
print("set s1= ",s1)
s.remove(3)
print("set s= ",s)
```

```python
# A frozen set
# Frozen sets in Python are immutable objects that only support methods
# and operators that produce a result without affecting the frozen set 
# or sets to which they are applied.
# While elements of a set can be modified at any time, 
# elements of the frozen set remain the same after creation. 

fs = frozenset(["e", "f", "g"])

print("Frozen Set")
print(fs)
# fs.add("h") # this gives an error
```
