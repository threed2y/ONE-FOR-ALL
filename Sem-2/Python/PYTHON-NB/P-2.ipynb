WRITE A PROGRAM TO TAKE A NUMBER N AS A INPUT AND FIND THE SUM OF FIRST N NATURAL NUMBERS.

```python
n = int(input("PLEASE GIVE ME A NUMBER OF YOUR CHOICE:"))
sum = (n*(n+1))/2
   
for n in range(n+1):
    print(n)
print("THE SUMMATION OF FIRST n NUMBERS IS : ",sum)
```

WRITE A PROGRAM WHICH GIVES OUTPUT OF SQUARE OF EACH NATURAL NUMBERS, AND FIND THE SUM OF THE SQUARES.

```python
r = int(input("PLEASE INPUT A NUMBER OF YOUR CHOICE:"))
sum_sq = r * (r + 1) * (2 * r + 1) // 6

for r in range(r+1):
    print(r,", THE SQUARE OF THIS NUMBER IS: " , r*r)
print("THE SUMMATION OF SQUARES OF FIRST n NATURAL NUMBERS IS: ", sum_sq)
```

```python
n = int(input("give me your number:"))
a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

CHECK FOR LEAP YEAR

```python
n = int(input("ENTER THE YEAR"))
if n in range(0,10000):
    print("VALID YEAR, MOVE AHEAD")
if n % 100 | n% 400 == 0:
    print("THIS IS A LEAP YEAR 1")
elif n % 4 == 0:
    print("THIS IS A LEAP YEAR BUT SHITTY")
else :
    print("THIS IS NOT A LEAP YEAR")
```

CHECK FOR EVEN OR ODD NUMBER UNDER N NUMBER

```python
n = int(input("GIVE ME A NUMBER OF YOUR CHOICE:"))
for n in range(1,n):
    if n % 2 == 0:
        print("THE NUMBER IS EVEN",n)
    else:
        print("THE NUMBER IS ODD",n)
```
