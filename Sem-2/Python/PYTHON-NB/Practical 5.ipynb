```python
# CREATE A DICTIONARY TO STORE THE KEY,VALUE PAIRS
keys = ["raja","alok","tanya","dipanshu","matra","rashmi","tiwari","aditya"]
values = (50,43,49,99,78,9,10,29)
student_record = {}

new_record = dict(zip(keys,values))
student_record.update(new_record)

max_marks = max(student_record,key=student_record.get)
print(max_marks, student_record[max_marks])
```

```python
# STORE STUDENT DATA

student = (("Name","Amit"),("Age",20),("Grades",85.5),("Course","BSc.Statistics"))

student = student + (("location","Vadodara"),)

dict_student = dict(student)
print("NAME:",dict_student.get("Name"))
print("AGE:",dict_student.get("Age"))
print("GRADES:",dict_student.get("Grades"))
print("Course:",dict_student.get("Course"))
print("Location:",dict_student.get("location"))
```

```python
#SET MANIPULATION
set_n = set()
set_k = set()
for i in range(5):
    n,k = map(int,(input().split()))
    set_n.add(n)
    set_k.add(k)

print("union:",set_k | set_n)
print("intersection:",set_k & set_n)
print("difference:",set_k-set_n)
```

```python
# STORE COURSE NAMES AS KEYS AND SET OF STUDENTS AS VALUES

directory = {}

while True:
    choice = int(input(
        "What do you want to do?\n"
        "1. Add student to course\n"
        "2. Remove student from course\n"
        "3. Search course\n"
        "4. Update student name in course\n"
        "5. Exit\n"
    ))

    if choice == 1:
        course = input("Enter course name: ")
        student = input("Enter student name: ")

        if course not in directory:
            directory[course] = set()
        directory[course].add(student)
        print(f"{student} added to {course}.")

    elif choice == 2:
        course = input("Enter course name: ")
        student = input("Enter student name: ")
        if course in directory and student in directory[course]:
            directory[course].remove(student)
            print(f"{student} removed from {course}.")
        else:
            print("Course or student not found.")

    elif choice == 3:
        course = input("Enter course name to search: ")
        if course in directory:
            print(f"Students in {course}: {directory[course]}")
        else:
            print("Course not found.")

    elif choice == 4:
        course = input("Enter course name: ")
        old_name = input("Enter current student name: ")
        new_name = input("Enter new student name: ")
        if course in directory and old_name in directory[course]:
            directory[course].remove(old_name)
            directory[course].add(new_name)
            print(f"{old_name} updated to {new_name} in {course}.")
        else:
            print("Course or student not found.")

    elif choice == 5:
        print("Exiting...")
        break

    else:
        print("Invalid choice.")

    print("Current directory:", directory)
```
