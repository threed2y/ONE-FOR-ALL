```python
#version - 3

def freq_dist(st):
      d=dict()
      for i in st:
          if i not in d:
              d[i] =1
          else:
              d[i] += 1
      return d

def print_freq_dist(d1):
    print('ch','freq')
    print('----------')
    for i in d1:
        print(i,"\t",d1[i])

# main program
s="Hi all ! @ This is a Program to output count of each character that appers in the given string **"
from string import punctuation
for i in punctuation:
    s=s.replace(i,'') # this will remove every punctuation character from string s 
s=s.replace(' ','')
S=s.lower()
print('new string is, S = ',S)
print('no. of characters in S:',len(S))
f=freq_dist(S)
print('the freq dist of the given data is:')
print_freq_dist(f)
```

```python
# Version - 4

def freq_dist(st):
      d=dict()
      for i in st:
          if i not in d:
              d[i] =1
          else:
              d[i] += 1
      return d

# function to invert a dictionary
def invert_dict(d): # d is a dictionary
    inverse=dict()
    for k in d:
        val=d[k]
        if val not in inverse:
            inverse[val]=[k]
        else:
            inverse[val].append(k)
    return inverse

# function to print the inverted dictionary in the tabular form
def print_freq_dist(d1):
    print('freq','     ch  ')
    print('----------------')
    for i in d1:
        print(i,"  ",d1[i])

# main program
s="Hi all ! @ This is a Program to output count of each character that appers in the given string **"
S=s.lower()
print('no. of characters in S:',len(S))
f=freq_dist(S)
print('the freq dist of the given data is:')
print(f)
print_freq_dist(f)
f1=invert_dict(f)
print('the inverted dictionary d is as follows:',f1 )
print('the inverted dictionary in the tabular form is as follows:')
print_freq_dist(f1)
```

```python
## functions having multiple parameters specified in different formats
## that is, sometimes we don't know exactly how many paramters will be 
# there in a function call
# The special syntax *args in function definitions in Python
# is used to pass a variable number of arguments to a function.
# It is used to pass a non-keyworded, variable-length argument list.  
#-----------------------------------------------------

def items(*things): # 'things' will be a tuple
    print(things)
basket1=('x',) # a tuple with a single element
basket2=('x','y','z')
items(*basket1)
items(*basket2)
items('a','b','c','d')
items('i1','i2','i3','i4','i5','i6')
```

```python
def profile(name,*ages): # ages can be a tuple with any number of values
    print("name of the employee:%s" %name)
    print("ages of employee and dependents if any:",ages)
a1=(42,40,13,10)
profile("p1",*a1)
profile('Person1',26)
profile('Person2',42,40,13,10)
profile('person3',35,31,5)
```

```python
The special syntax **kwargs in function definitions
in Python is used to pass a keyworded, variable-length 
argument list. We use the name kwargs with the double star.
The reason is that the double star allows us to pass 
through keyword arguments (and any number of them). 
```

```python
def family(**kwargs):  
    for key, value in kwargs.items(): 
        print ("% s = % s" %(key, value))
        
print("\nResult of **kwargs for family-1")
family(member1 ='X', member2 ='Y')

print("\nResult of **kwargs for family-2")
family(member1 ='X', member2 ='Y', member3='Z')
```

```python
# name and age of family members
def new_family(**kwargs):  
    for key, value in kwargs.items(): 
        print ("% s = % d" %(key, value))
       

print("\nResult of **kwargs for family-1")
new_family(husband =35, wife =33)

print("\nResult of **kwargs for family-2")
new_family(husband =40, wife =39, kid1=13, kid2=10)

d={'a': 10, 'b': 20, 'c': 30}
print("\nResult of **kwargs for dictionary d:")
new_family(**d)
```
