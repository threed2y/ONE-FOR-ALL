```python
import matplotlib.pyplot as plt

# Load the digits dataset from datasets module
from sklearn import datasets
digits = datasets.load_digits()
```

```python
print(digits.DESCR)
```

```python
X = digits.data
Y = digits.target
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, 
                                                    test_size=0.33, 
                                                    random_state = 87779 )
```

## Bayesian Classification Digit dataset


```python
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
# import Gaussian Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB

# create a Gaussian Classifier
classifer = GaussianNB()

# training the model
classifer.fit(X_train, y_train)

# testing the model
y_pred = classifer.predict(X_test)
```

```python

print('Accuracy = %0.4f'%np.mean(y_pred == y_test))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(y_test, y_pred))
```

```python
# importing classifier
from sklearn.naive_bayes import BernoulliNB

# initializaing the NB
classifer1 = BernoulliNB()

# training the model
classifer1.fit(X_train, y_train)

# testing the model
y_pred1 = classifer1.predict(X_test)
```

```python
print('Accuracy = %0.4f'%np.mean(y_pred1 == y_test))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(y_test, y_pred1))
```
