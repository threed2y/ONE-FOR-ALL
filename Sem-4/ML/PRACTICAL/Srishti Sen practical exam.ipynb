Ques . The breast_cancer dataset is the Wisconsin Diagnostic Breast Cancer (WDBC) dataset. The dataset consists of data on the characteristics of the cell nuclei found in the image of a fine needle aspirate of a breast mass.

```python
from sklearn import datasets
#importing breast cancer dataset
data=datasets.load_breast_cancer()
```

```python
print(data.DESCR)
```

```python
data.feature_names
```

```python
data.target_names
```

```python
X=data.data
Y=data.target
```

```python
# Train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33)
```

Use this data as a training data for classifying the tumor as Malignant or Benign using logisticregression. To build the model, select the suitable features from the available features using graphical exploration. Output the learned model along suitable performance as measured using cross validation method.

```python
import matplotlib.pyplot as plt
plt.boxplot(X)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(penalty="none", solver="lbfgs")
logr.fit(X_train, Y_train)
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score 
import numpy as np
Y_pred = logr.predict(X_test)
```

```python
#Cross validation method
cv_scores=cross_val_score(logr, X, Y, cv = 5)
print('Accuracy =',np.mean(cv_scores))
print("Confusion matrix:",metrics.confusion_matrix(Y_test, Y_pred))
```
