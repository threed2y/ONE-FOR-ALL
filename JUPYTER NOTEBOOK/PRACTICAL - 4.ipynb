```python
# 1. WRITE A PROGRAM TO TAKE A POSITIVE INTEGER FROM THE USER AND COUNTS DOWN TO ZERO.
import time

n = int(input())
current_time = time.time() # FOR TIME WITH PRINT VALUES

num = n
while num > 0:
     if num > 100: # IF THE CURRENT VALUE IS GREATER THAN 100, DECREASE BY 10 IN EACH ITERATION
         num -= 10
         print(num,current_time)
         
     elif num <= 100: # IF THE CURRENT VALUE IS LESS THAN OR EQUAL TO 100, DECREASE BY 1 IN EACH ITERATION.
         num -= 1
         print(num,current_time)
```

```python
# 2. FREQUENCY DISTRIBUTION OF EACH STRING

char = str(input()) 
freq_map = {}

for i in range(len(char)):
    freq_map[char[i]] = freq_map.get(char[i],0) + 1

print(freq_map)

```

```python
# 3. SEPARATING A STRING OF DIFFERENT CHARACTERS

inp = "A1$b2_C3d4*E5f6#G7h8!9@0"

s = inp.lower()
char = []
num = []
alpha = []

s = sorted(s)

for i in s:
    if i.isdigit()  :
        num.append(i)
    elif i.isalpha():
        alpha.append(i)
    else:
        char.append(i)
print("char:",char,"num:",num,"alpha:",alpha)
```

```python
# 5. SUBSEQUENCE OF A STRING

s1 = "Students learn Python and improves Skills in Technology"
s2 = "Python is a powerful programming language used for Data Analysis and Machine Learning. Many students learn python to build application, analyze data, and solve real world problems. learning python improves logical thinking, coding skills, and career oppurtunities in technology."

s2 = s2.lower()
s1 = s1.lower()

i,j = 0,0

while i < len(s2) and j < len(s1):
    if s2[i] == s1[j]:
        j += 1
    i += 1

sub = (j == len(s1))
print(sub)
```
